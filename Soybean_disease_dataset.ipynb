```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# Define the list of Soybean links
soybean_links = [
    'https://soybeanresearchinfo.com/soybean-disease/septoria-brown-spot/',
    'https://soybeanresearchinfo.com/soybean-disease/bacterial-blight/',
    'https://soybeanresearchinfo.com/soybean-disease/bacterial-pustule/',
    'https://soybeanresearchinfo.com/soybean-disease/frogeye-leaf-spot/',
    'https://soybeanresearchinfo.com/soybean-disease/cercospora-leaf-blight/',
    'https://soybeanresearchinfo.com/soybean-disease/downy-mildew/',
    'https://www.bhg.com/how-to-grow-soybean-plants-7371324'
]

# Function to scrape and process a Soybean link
def scrape_soybean_link(link):
    try:
        # Send an HTTP GET request to the Soybean page
        response = requests.get(link)
        response.raise_for_status()

        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract text content from the page and remove HTML tags
        text = soup.get_text()

        # Remove special characters and multiple spaces
        text = re.sub(r'\s+', ' ', text)
        text = re.sub(r'[^\w\s]', '', text)

        # Split the text into 50-word chunks with 10-word overlaps
        chunk_size = 50
        overlap = 10
        chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size - overlap)]

        return chunks
    except Exception as e:
        print(f"Error scraping {link}: {str(e)}")
        return []

# Create an empty DataFrame to store the data
df = pd.DataFrame(columns=['Link', 'Chunk'])

# Iterate through the Soybean links and scrape/process each one
for link in soybean_links:
    chunks = scrape_soybean_link(link)

    # Add the chunks to the DataFrame
    for chunk in chunks:
        df = df.append({'Link': link, 'Chunk': chunk}, ignore_index=True)

# Display the first few rows of the DataFrame
print(df.head())

# Save the DataFrame to a CSV file if needed
df.to_csv('soybean_chunks.csv', index=False)
```
